```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import requests
from keras.utils import to_categorical
np.random.seed(10)
```

```python
# Importing and cleaning data using pandas library
data = pd.read_csv('C_data.csv')
del data['Unnamed: 32']
```

```python
## Observe the data
data
```

```python
## We have left first two columns and taken other columns as input features
X = data.iloc[:, 2:].values

# 2nd column is output labels
y = data.iloc[:, 1].values
```

```python
## Convert the output labels to numbers : M->0; B-> 1
## Store the output in Y_v
Y_v = []

# YOUR CODE HERE
for i in y:
  if i=='B':
    Y_v.append(1)
  else:
    Y_v.append(0)
```

```python
### One-hot encode Y_v
def oneHot(y, Ny):
    '''
    Input:
        y: an int in {0, 1}
        Ny: Number of classes, e.g., 2 here.
    Output:
        Y: a vector of Ny (=2) tuples
    '''
    # YOUR CODE HERE
    return to_categorical(y, num_classes = Ny)
    
```

```python
from keras.utils import to_categorical

Ny = len(np.unique(Y_v)) # Ny is number of categories/classes


Y_v = to_categorical(Y_v[:], num_classes = Ny) # 
```

```python
### Split data into train and test. Keep 10% of samples for testing
## Divide the data into these variables - X_train, X_test, y_train, y_test
# YOUR CODE HERE
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y_v, test_size = 0.10, random_state = 1)
```

```python
"""test for splitting"""
assert(X_train.shape[0] == 512)
```

```python
## Normalize the Data
def findMeanStddev(X):
    '''
    Input: 
        X: a matrix of size (no. of samples, dimension of each sample)
    Output:
        mean: mean of samples in X (same size as X)
        stddev: element-wise std dev of sample in X (same size as X)
    '''
    # YOUR CODE HERE
    mean = X.mean(axis=0) 
    stddev = X.std(axis=0)
    return mean , stddev
 
def normalizeX(X, mean, stddev):
    '''
    Input:
        X: a matrix of size (no. of samples, dimension of each sample)
        mean: mean of samples in X (same size as X)
        stddev: element-wise std dev of sample in X (same size as X) 
    Output:
        Xn: X modified to have 0 mean and 1 std dev
    '''
    # YOUR CODE HERE
    X_ = (X - mean)/stddev
    return X_
```

```python
#Normalize X_train data
mean, stddev = findMeanStddev(X_train)
X_train=normalizeX(X_train, mean, stddev)
mean, stddev = findMeanStddev(X_test)
X_test=normalizeX(X_test, mean, stddev)

```

```python
import keras
# YOUR CODE HERE
from keras.models import Sequential
from keras.layers import Dense
#create model
model = Sequential()



#add model layers

model.add(Dense(10, input_dim=X_train.shape[1], activation='sigmoid'))
model.add(Dense(20, activation='sigmoid'))
model.add(Dense(10, activation='sigmoid'))
model.add(Dense(Ny, activation='softmax')) # 'sigmoid'
# Compile model
```

```python
# Compile model
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy']) # 'binary_crossentropy', 'adam', 'adagrad', 'sgd'
```

```python
model.summary()
```

```python
# Fit the model
model.fit(X_train, y_train, epochs=100, batch_size=100)
```

```python
# calculate predictions
predictions = model.predict(X_train)
# round predictions
rounded = [round(x[0]) for x in predictions]
print(rounded)
```

```python
"""Test for model"""
```

```python
# evaluate the model
scores = model.evaluate(X_test, y_test)
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```

```python
from sklearn.metrics import classification_report, confusion_matrix  
predictions = np.argmax(model.predict(X_test), axis=1)
y_test = np.argmax(y_test,axis=1)
print(confusion_matrix(y_test,predictions))  
print(classification_report(y_test,predictions))  
```
